```python
#set all the paths
source_folder = '/content/gdrive/MyDrive/TFE_crowd_counting/Dataset/Basic datas/images/Test/'
destination_folder =  '/content/gdrive/MyDrive/TFE_crowd_counting/Dataset/Basic datas/images/Test_small/'
new_GT_folder = '/content/gdrive/MyDrive/TFE_crowd_counting/Dataset/Basic datas/ground_truth/Test/'
GT_folder = '/content/gdrive/MyDrive/TFE_crowd_counting/Dataset/Basic datas/ground_truth/Test_small/'

#Load new images from the folder
images = []
GTs = []
source_list = sorted(os.listdir(source_folder))
GT_source_list = sorted(os.listdir(new_GT_folder))

for filename in source_list:
  img = cv2.imread(os.path.join(source_folder,filename))
  if img is not None:
    images.append(img)

for filename in GT_source_list:
  if not (new_GT_folder+filename).endswith('.mat'):
    continue
  else:
    GT= io.loadmat(new_GT_folder+filename)
    if GT is not None:
      GTs.append(GT)

print(len(images))
print(len(GTs))
```

```python
print(images[0].shape)
```

```python
def full_resize(images,dims):
    for ind in range(len(images)):
      img = images[ind]
      GT = GTs[ind]

      #resize and save image
      resized = cv2.resize(img, dims, interpolation = cv2.INTER_AREA)
      filename = destination_folder+'IMG_small_'+str(ind+1)+'.png'
      cv2.imwrite(filename, resized)

      #resize and save ground_truth
      ratio_x = img.shape[0]/resized.shape[0]
      ratio_y = img.shape[1]/resized.shape[1]

      information = GT['image_info']
      i=information[0][0]
      x=i.item(0)
      for j in range(len(x[0])):
        x[0][j][0]= x[0][j][0]/ratio_x
        x[0][j][1]= x[0][j][1]/ratio_y
      io.savemat(GT_folder+'IMG_small_'+str(ind+1)+'.mat',GT)
```
